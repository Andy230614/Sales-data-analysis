```python
df.duplicated().sum()
df = df.drop_duplicates()
```

```python
df.dtypes
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"])
```

```python
df["Profit Margin"] = df["Profit"]/df["Sales"]
df.sort_values("Profit Margin").head(10)
```

What is the total sales?

```python
Total_sales=df['Sales'].sum()
print(Total_sales)
```

How much total revenue did the company generate?

```python
Total_sales=df['Sales'].sum()
print('Total_sales:', Total_sales)
```

```python
Total_profit=df['Profit'].sum()
print('Total_profit:',Total_profit)
```

Sales by Region

```python
Sales_by_region=df.groupby('Region')['Sales'].sum()
print('Sales_by_region:', Sales_by_region)
```

```python
import matplotlib.pyplot as plt

Sales_by_region.plot(kind="bar")

plt.title('Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')

plt.savefig("../Images/sales_by_region.png")

plt.show()
```

Profit by Category

```python
Profit_by_category=df.groupby('Category')['Profit'].sum()
print('Profit_by_category:', Profit_by_category)
```

```python
Profit_by_category.plot(kind='bar')

plt.title('Profit by Category')
plt.xlabel('Category')
plt.ylabel('Profit')

plt.savefig("../Images/profit_by_category")
plt.show()
```

Top 10 selling products

```python
Top_products=df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
print('Top_product:', Top_products)
```

Monthly Sales Trend

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Month'] = df['Order Date'].dt.to_period('M')
```

```python
Monthly_sales = df.groupby('Month')['Sales'].sum()

Monthly_sales.plot()

plt.title('Monthly Sales')
plt.xlabel('Month')
plt.ylabel('Sales')

plt.savefig("../Images/monthly_sales")
plt.show()
```

Sales by Segment

```python
Sales_by_segment = df.groupby("Segment")["Sales"].sum()
print("Sales by Segments:", Sales_by_segment)
```

```python
Sales_by_segment.plot(kind='bar')

plt.title("Sales by Segment")
plt.xlabel("Segment")
plt.ylabel('Sales')

plt.savefig("../Images/sales_by_segment")

plt.show()
```
